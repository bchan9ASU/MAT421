```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

Linear Space

A linear space (vector space) is a collection of vectors that can be added together and scaled by numbers.

A vector space satisfies:

Closure under addition: If u and v are in the space, then u + v is also in the space.

Closure under scalar multiplication: If v is in the space and c is a scalar, then c * v is also in the space.

Below, we create a simple example of a 2D linear space.

```python
# Define two basis vectors
v1 = np.array([1, 2])
v2 = np.array([2, -1])

# Generate linear combinations of the basis vectors
scalars = np.linspace(-3, 3, 10)
vectors = [a * v1 + b * v2 for a in scalars for b in scalars]

# Convert to array for plotting
vectors = np.array(vectors)

# Plot the linear space
plt.figure(figsize=(6, 6))
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.axvline(0, color='black', linestyle='--', linewidth=1)
plt.scatter(vectors[:, 0], vectors[:, 1], color='blue', alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Linear Space Example')
plt.grid()
plt.show()
```

Orthogonality

Two vectors u and v are orthogonal if their dot product is zero:

u ⋅ v = 0

This means they are at a 90-degree angle to each other.

Below, we check if two given vectors are orthogonal.

```python
# Define two vectors
u = np.array([1, 2])
v = np.array([-2, 1])

# Compute dot product
dot_product = np.dot(u, v)

# Check orthogonality
if dot_product == 0:
    print("The vectors are orthogonal.")
else:
    print("The vectors are NOT orthogonal.")
```

Eigenvalues and Eigenvectors

For a square matrix A, an eigenvector v and an eigenvalue λ satisfy:

A * v = λ * v

Eigenvalues give important information about transformations represented by the matrix.

Below, we compute the eigenvalues and eigenvectors of a matrix.

```python
# Define a matrix
A = np.array([[4, -2],[1,  1]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Display results
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

Linear Regression

Linear regression models the relationship between a dependent variable y and one or more independent variables X using:

y = mX + b

where:

m is the slope (coefficient)

b is the intercept

Below, we fit a simple linear regression model using Python.

```python
# Generate synthetic data
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Feature values
y = 3 * X + 7 + np.random.randn(100, 1) * 2  # Linear relation with noise

# Train a linear regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plot results
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', linewidth=2, label='Linear Regression Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Example')
plt.legend()
plt.grid()
plt.show()

# Print model parameters
print("Slope (m):", model.coef_[0][0])
print("Intercept (b):", model.intercept_[0])
```
