```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d, CubicSpline
```

This Jupyter Notebook demonstrates interpolation techniques, including linear interpolation
and cubic spline interpolation. Examples are chosen to effectively explain the concepts.
Interpolation involves estimating unknown values that fall between known data points.
It's used extensively in numerical analysis, data science, and engineering.

The interpolation problem aims to estimate the value of a function between two known data points. Given
independent data values x and dependent data values y, where the relationship is known for a discrete
set of points, interpolation estimates y values for any x within the range of the dataset. Below is
a visualization of a simple dataset and the interpolation problem.

```python
# Sample Dataset
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 2, 1, 3, 7, 8])

# Visualization of the Problem
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o', label='Known Data Points')
plt.title('Interpolation Problem Statement')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

Linear interpolation assumes that the estimated point lies on the straight line joining the nearest
data points to the left and right.

```python
# Create linear interpolator
linear_interp = interp1d(x, y)

# Generate new x values for interpolation
x_new = np.linspace(min(x), max(x), 100)

# Interpolated values
y_linear = linear_interp(x_new)

# Plot linear interpolation
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o', label='Known Data Points')
plt.plot(x_new, y_linear, '-', label='Linear Interpolation')
plt.title('Linear Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

Cubic spline interpolation creates a piecewise cubic polynomial that passes through each of the
known data points, ensuring smoothness at the boundaries.

```python
# Create cubic spline interpolator
cubic_spline = CubicSpline(x, y)

# Interpolated values
y_cubic = cubic_spline(x_new)

# Plot cubic spline interpolation
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o', label='Known Data Points')
plt.plot(x_new, y_cubic, '-', label='Cubic Spline Interpolation')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
